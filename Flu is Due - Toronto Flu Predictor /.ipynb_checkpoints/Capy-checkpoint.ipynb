```python
import pandas as pd 

templated = pd.read_csv('templated.csv')
templated.head()
```

```python
i=templated['institution']

sunny = templated[
    (templated['institution'] == 'sunnybrook health sciences centre') & 
    (templated['causative1'] == 'Coronavirus')
]

corona = sunny[['begin_date', 'end_date']]
corona['infection'] = 1
corona
```

```python
sunny_corona = pd.concat([
    pd.DataFrame({
        'start': pd.date_range(row.begin_date, row.end_date, freq='D'),
        'infection': row.infection
    })
    for i, row in corona.iterrows()], 
    ignore_index=True)
```

```python
calendar = pd.DataFrame({'start': pd.date_range(start='2016-01-01', end='2018-11-30')})
```

```python
calendar.head()
```

```python
sunny_corona.head()
```

```python
s1 = pd.merge(calendar, sunny_corona, how='left', on=['start'])
```

```python
s1 = s1.fillna(0)
```
